```python
import cv2
import matplotlib.pyplot as plt
```

```python
def apply_gaussian_blur(image_path):
    # Membaca citra
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Mengubah ke RGB

    # Menerapkan Gaussian Blur
    blurred_image = cv2.GaussianBlur(image, (11, 11), 0)
    
    # Menampilkan citra asli dan citra yang telah dihaluskan
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(image), plt.title('Original Image')
    plt.xticks([]), plt.yticks([])
    plt.subplot(122), plt.imshow(blurred_image), plt.title('Blurred Image')
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
apply_gaussian_blur('Reina.jpg')
```
